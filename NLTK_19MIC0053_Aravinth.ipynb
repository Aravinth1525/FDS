```python
from nltk.corpus import names

def gender_features(word):
  return{'last_letter':word[-1]}
```

```python
gender_features('Donald')
```

```python
labeled_names=([(name,'male') for name in names.words('male.txt')]+[(name,'female') for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labeled_names)
```

```python
featuresets=[(gender_features(n),gender) for (n,gender) in labeled_names]
```

```python
train_set,test_set=featuresets[500:],featuresets[:500]
```

```python
classifier=nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('Donald'))
```

```python
classifier.classify(gender_features('Pranauv'))
```

```python
texts=""" Jai Bhim (alternatively spelled Jay Bhim or Jai Bheem; transl."Victory for Bhim", "Long live Bhim" or "Hail Bhim") is a slogan and greeting used by followers of Bhimrao Ramji Ambedkar, an Indian scholar, social reformer and the chief architect of the Constitution of India. Jai Bhim literally means "Victory to Bhim" referring to B. R. Ambedkar.[1][2] In 1935, Jai Bhim conceived and developed by Babu L. N. Hardas (1904  1939),[3][4] a staunch follower of Babasaheb Ambedkar, and the secretary of Samata Sainik Dal.[5][6] However, its origin is obscure and may date as far back as 1818, well before Ambedkar's birth. """
```

```python
for text in texts:
  sentences=nltk.sent_tokenize(texts)
  for sentence in sentences:
    words=nltk.word_tokenize(sentence)
    tagged=nltk.pos_tag(words)
    print(tagged)
```
